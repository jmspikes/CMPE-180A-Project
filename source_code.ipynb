```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
# import neccesary libraries 
import numpy as np
from tabulate import tabulate
import sys
import seaborn as sns

# US Airline Flight Routes and Fares
# https://www.kaggle.com/code/nitikagupta29/us-airline-flight-routes-and-fares

#  uncomment for google drive
#  filename = "/content/drive/MyDrive/data/US Airline Flight Routes and Fares 1993-2024.csv"
filename = 'US_Airline_Flight_Routes_and_Fares_1993-2024.csv'

# df = pd.read_csv(filename)
df = pd.read_csv(filename, low_memory=False)

df.head()


# Load flight data
flight_data = pd.read_csv('C:\Users\yakuz\PycharmProjects\CMPE-180A-Project\US_Airline_Flight_Routes_and_Fares_1993-2024.csv')

# Load airport data
airport_data = pd.read_csv('airports.csv')

# Merge flight data with airport data to get lat/long
flight_data = flight_data.merge(airport_data, left_on='origin', right_on='airport_code')
flight_data = flight_data.merge(airport_data, left_on='destination', right_on='airport_code', suffixes=('_origin', '_destination'))

# Load USA map
usa = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
usa = usa[usa.name == "United States of America"]

# Plotting function
def plot_flight_data(year):
    yearly_data = flight_data[flight_data['year'] == year]
    
    fig, ax = plt.subplots(1, 1, figsize=(15, 15))
    usa.plot(ax=ax, color='white', edgecolor='black')
    
    for _, row in yearly_data.iterrows():
        plt.plot([row['longitude_origin'], row['longitude_destination']],
                 [row['latitude_origin'], row['latitude_destination']],
                 linewidth=row['frequency'] / 10, color='blue')  # Adjust linewidth scaling as needed
    
    plt.title(f'Flight Data for {year}')
    plt.show()

# Plot for each year
for year in range(1994, 2023 + 1):
    plot_flight_data(year)
```
